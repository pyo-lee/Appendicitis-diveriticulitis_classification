```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np 
import os
import glob
#import cv2
#from libtiff import TIFF

def makeFolder():
    # make folder
    folder = []
    split_num =5
    for i in range(1, split_num+1):
        folder = []
        folder.append('D:/appendicitis/data/3class/train')
        folder.append('D:/appendicitis/data/3class/test')
        folder.append('D:/appendicitis/data/3class/train/appendicitis')
        folder.append('D:/appendicitis/data/3class/train/no_lesion')
        folder.append('D:/appendicitis/data/3class/train/diverticulitis')
        folder.append('D:/appendicitis/data/3class/test/appendicitis')
        folder.append('D:/appendicitis/data/3class/test/no_lesion')
        folder.append('D:/appendicitis/data/3class/test/diverticulitis')
        folder.append('D:/appendicitis/data/3class/npy')

        for j in range(len(folder)):
            if not os.path.lexists(folder[j]):
                os.mkdir(folder[j])

class myAugmentation(object):

    """
    A class used to augmentate image
    Firstly, read train image and label seperately, and then merge them together for the next process
    Secondly, use keras preprocessing to augmentate image
    Finally, seperate augmentated image apart into train image and label
    """

    def __init__(self, 
                 train_ap_path="C:/Users/gcubme/Desktop/eye/atrophy", 
                 aug_path="C:/Users/gcubme/Desktop/eye/atrophy", 
                 img_type="jpg"):

        """
        Using glob to get all .img_type form path
        """

        self.train_ap_imgs = glob.glob(train_ap_path+"/*."+img_type)
        self.train_ap_path = train_ap_path
        self.aug_path = aug_path
        self.img_type = img_type
        
        self.slices = len(self.train_ap_imgs)
        self.datagen = ImageDataGenerator(
                                    rotation_range=5,
                                    width_shift_range=0.05,
                                    height_shift_range=0.05,
                                    shear_range=0.05,
                                    zoom_range=0.05,
                                    horizontal_flip=True,
                                    vertical_flip=True,
                                    fill_mode='nearest')

    def Augmentation(self):

        
        #ap data augmentation

        trains_ap = self.train_ap_imgs
        path_train_ap = self.train_ap_path
        path_merge_ap = self.aug_path
        imgtype = self.img_type
        if len(trains_ap) == 0 or len(trains_ap) == 0:
            print ("trains can't match labels")
            return 0
        imgs = glob.glob(path_train_ap+"/*."+self.img_type)
        k=0        
        for imgname in imgs:
            print(imgname)                
            midname = imgname[imgname.rindex("\\")+1:] 
            img_t = load_img(path_train_ap +"/" + midname)
            #img_tmp = img_t
            img = img_to_array(img_t)  
            #img_tmp.save(path_merge_ap+"/"+str(k)+"."+imgtype)
            #img = x_t
            img = img.reshape((1,) + img.shape)
            savedir = path_merge_ap# + "/" + str(k)
            if not os.path.lexists(savedir):
                os.mkdir(savedir)
            self.doAugmentate(img, savedir, str(k))
            k=k+1


    def doAugmentate(self, img, save_to_dir, save_prefix, batch_size=1, save_format='png', imgnum=18):

        datagen = self.datagen
        i = 0
        for batch in datagen.flow(img, batch_size=batch_size,save_to_dir=save_to_dir,save_prefix=save_prefix,save_format=save_format):

            i=i+1
            if i > imgnum:
                break           



if __name__ == "__main__":

# 	makeFolder()  
	aug = myAugmentation()  
	aug.Augmentation()
	##aug.splitMerge()
	##aug.splitTransform()

	#mydata = dataProcess(256,256)
	#for index in range(1,6):
# 		print('exp%d strat'%index)
# 		mydata.create_train_data(index)
# 		mydata.create_test_data(index)
# 		print('exp%d done'%index)
# 		print('*'*30)
# 	#imgs_train,imgs_mask_train = mydata.load_train_data()  
	#imgs_test,imgs_mask_test = mydata.load_test_data()
```
