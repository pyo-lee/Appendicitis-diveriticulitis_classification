```python
import os
import cv2
import pydicom
import math
import pandas as pd
```

```python
path='D:/appendicitis/1_data_raw/data/No_lesion_png_2'
p_id=os.listdir(path)
label='no_lesion'
result=pd.DataFrame(columns=['patient_id','label','slice_num','row_length','colum_length','effective_diameter'])

for m in range(len(p_id)):
    img_name=os.listdir(path+'/'+p_id[m])
    save_path='D:/appendicitis/7_effective_diameter/img_no_lesion/'+p_id[m]
#     os.mkdir(save_path)
    l=0
    for n in range(len(img_name)):
        slice_name=img_name[n][9:13] 
        img=cv2.imread(path+'/'+p_id[m]+'/'+img_name[n])
        imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret,thresh= cv2.threshold(imgray,1,255, cv2.THRESH_BINARY)
        contours,_= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
        #image=cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
        
        tmp=[]
        for j in range(len(contours)):
            cnt = contours[j]
            mmt = cv2.moments(cnt) 
            tmp.append(float(mmt['m00']))
        max_num = max(tmp)
        index = tmp.index(max_num)
        #print(index)
        contour=contours[index]
        image=cv2.drawContours(img, contour, -1, (0, 255, 0), 3)
        
        #########(x,y)###########
        xy_list=[]
        a=contour.tolist()
        x_list, y_list =[], []

        for i in range(len(a)):
            x, y = a[i][0]
            y_list.append(y)
            x_list.append(x)

        for i in range(len(x_list)):
            xy = x_list[i], y_list[i]
            #xy.tolist()
            xy_list.append(xy)
        
        #######x축#########
        same_y = []
        n_y = set(y_list)
        for i in n_y:
            tmp=[]
            y_tmp=[]
            for k in range(1,len(xy_list)):
                if i == xy_list[k][1]:
                    #print(xy_list[k])
                    if not i in tmp:
                        #y_tmp.append(xy_list[k][1])
                        tmp.append(xy_list[k])    
                    
            same_y.append(tmp)
            if [] in same_y:
                same_y.remove([])
        diff=[]
        for i in range(len(same_y)):
            max_x=max(same_y[i])
            min_x=min(same_y[i])
            diff.append(max_x[0]-min_x[0])
            
        index=diff.index(max(diff))
        index_x1=xy_list.index(same_y[index][0])
        index_x2=xy_list.index(same_y[index][-1])
        
        ###########y축################
        same_x = []

        n_x = set(x_list)

        for i in n_x:
            tmp=[]
            x_tmp=[]
            for k in range(1,len(xy_list)):
                if i == xy_list[k][0]:
                    #print(xy_list[k])
                    if not i in tmp:
                        #y_tmp.append(xy_list[k][1])
                        tmp.append(xy_list[k])            
            same_x.append(tmp)
            if [] in same_y:
                same_x.remove([])

        diff=[]
        for i in range(len(same_x)):
            max_y=max(same_x[i])
            min_y=min(same_x[i])
            diff.append(max_y[1]-min_y[1])

        index=diff.index(max(diff))
        print(max(diff), index)
        index_y1=xy_list.index(same_x[index][0])
        index_y2=xy_list.index(same_x[index][-1])
        
        image_r=cv2.line(image, xy_list[index_x1], xy_list[index_x2], (0,255,225),2)
        image_r=cv2.line(image_r, xy_list[index_y1], xy_list[index_y2], (255,0,0),2)
        
        ##############################################
        if not os.path.isdir(save_path):
            os.mkdir(save_path)
        #cv2.imwrite(save_path+'/'+img_name[n],image_r)
        ##############################################
        
        dcm_path='D:/appendicitis/1_data_raw/data/No_lesion'
        dcm_foler=os.listdir(dcm_path)
        dcm_name=os.listdir(dcm_path+'/'+dcm_foler[m])
        print(dcm_path+'/'+dcm_foler[m]+'/'+img_name[n][:-4]+'.dcm')
        try:
            ds = pydicom.read_file(dcm_path+'/'+dcm_foler[m]+'/'+img_name[n][:-4]+'.dcm')
        except:
            ds = pydicom.read_file(dcm_path+'/'+dcm_foler[m]+'/'+img_name[0][:8]+img_name[0][9:-4]+'.dcm')
        pixel_spacing=ds[0x0028, 0x0030].value
        
        row=abs((xy_list[index_x1][0]-xy_list[index_x2][0])*pixel_spacing[0])
        colum=abs((xy_list[index_y1][1]-xy_list[index_y2][1])*pixel_spacing[1])
        ed=math.sqrt(row*colum)
        if l==0:
            result=result.append({'patient_id':p_id[m],'label':label,'slice_num':str(slice_name),'row_length':str(row),'colum_length':str(colum),'effective_diameter':str(ed)},ignore_index=True)
        else:
            result=result.append({'patient_id': '','label':label,'slice_num':str(slice_name),'row_length':str(row),'colum_length':str(colum),'effective_diameter':str(ed)},ignore_index=True)
        l+=1
result.to_csv('D:/appendicitis/7_effective_diameter/ef_no.csv',index=False)
```

```python
path='D:/appendicitis/1_data_raw/data/Appendicitis_png_2/'
p_id=os.listdir(path)
label='appendicitis'
result=pd.DataFrame(columns=['patient_id','label','slice_num','row_length','colum_length','effective_diameter'])

for i in range(1):
    img_name=os.listdir(path+'/'+p_id[i])
    save_path='D:/appendicitis/7_effective_diameter/img/'+p_id[i]
    for k in range(len(img_name)):
        slice_name=img_name[k][9:13] 
        img=cv2.imread(path+'/'+p_id[i]+'/'+img_name[k])
        imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret,thresh= cv2.threshold(imgray,1,255, cv2.THRESH_BINARY)
        contours,_= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
        #image=cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
        
        tmp=[]
        for m in range(len(contours)):
            cnt = contours[m]
            mmt = cv2.moments(cnt) 
            tmp.append(float(mmt['m00']))
        max_num = max(tmp)
        index = tmp.index(max_num)
        #print(index)
        contour=contours[index]
        image=cv2.drawContours(img, contour, -1, (0, 255, 0), 3)
        
        x,y,w,h=cv2.boundingRect(contour)
        image_r=cv2.rectangle(image, (x,y), (x+w,y+h), (0,0,225),3)
        ##############################################
        if not os.path.isdir(save_path):
            os.mkdir(save_path)
        cv2.imwrite('D:/appendicitis/7_effective_diameter/img/'+p_id[i]+'/'+img_name[k],image_r)
        ##############################################
        
        dcm_path='D:/appendicitis/1_data_raw/data/Appendicitis/'
        dcm_foler=os.listdir(dcm_path)
        dcm_name=os.listdir(dcm_path+'/'+dcm_foler[i])
        print(dcm_path+'/'+dcm_foler[i]+'/'+img_name[k][:-4]+'.dcm')
        ds = pydicom.read_file(dcm_path+'/'+dcm_foler[i]+'/'+img_name[k][:-4]+'.dcm')
        pixel_spacing=ds[0x0028, 0x0030].value
        
        row=pixel_spacing[0]*w
        colum=pixel_spacing[1]*h
        ed=math.sqrt(row*colum)
        result=result.append({'patient_id':p_id[i],'label':label,'slice_num':str(slice_name),'row_length':str(row),'colum_length':str(colum),'effective_diameter':str(ed)},ignore_index=True)

result.to_csv('D:/appendicitis/7_effective_diameter/ef.csv',index=False)
```
