```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,roc_curve,auc
```

```python
df = pd.read_excel('./8_appendicitis_exp3/2class/prob.xlsx')
```

```python
df
```

```python
ori_gt = df['ori_GT'].tolist()
ori_pred = df['ori_pred'].tolist()
```

```python
cm=confusion_matrix(ori_gt,ori_pred)
print(cm)
print(cm.ravel())
tp,fn,fp,tn=cm.ravel()

print(tp,fn,fp,tn)
acc=accuracy_score(ori_gt,ori_pred)
sensitivity=tp/(tp+fn)
specificity=tn/(tn+fp)
precision=tp/(tp+fp)
recall=tp/(tp+fn)
F1_score=2*precision*recall/(precision+recall)

print(acc)
print(sensitivity)
print(specificity)
print(precision)
print(recall)
print(F1_score)
```
