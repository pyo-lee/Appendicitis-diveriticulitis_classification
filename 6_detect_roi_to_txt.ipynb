```python
import os
import glob
import numpy as np
from read_roi import read_roi_zip
import zipfile

png_path = './data/train_diverticulitis_png/' # png 파일 폴더
roi_path= './data/train_diverticulitis_roi/' # roi 파일 폴더
file_extention = '.png' # 파일 확장자
file_png_path = [] # jpg 저장할 환자 폴더
roi_zip_path=[]
png_file_name = os.listdir(png_path) # 환자 폴더
roi_file_name = os.listdir(roi_path)
for i in range(len(png_file_name)):
    name = png_path + png_file_name[i]+'/'
    file_png_path.append(name)
    roi_zip = roi_path + png_file_name[i] + '.zip'
    roi_zip_path.append(roi_zip)
# print(len(file_dcm_path), file_dcm_path)
# print(len(roi_zip_path),roi_zip_path)

zip_list_all = [] # roi 이름 저장

for i in range(len(roi_zip_path)):
    zip_file = zipfile.ZipFile(roi_zip_path[i])
    zip_list = zip_file.namelist()
    zip_list.sort()
    zip_list_all.append(zip_list)
a=0
c=1
f=open('./keras-retinanet-master/keras_retinanet/bin/train_diverticulitis.txt','w')
for i in range(len(file_png_path)):
    #print(c)
    for k in range(len(zip_list_all[i])):
        check=glob.glob(png_path+png_file_name[i]+'/'+'*.png')
        img=check
        roi_name = zip_list_all[i][k][:4]
        png_name = str(int(roi_name)-1).zfill(4)
        img_path = file_png_path[i]+png_file_name[i]+'_'+png_name+'.png'
        #print(img_path)
        #print(img)
        if img_path in img:
            #print('*****************************')
            roi = read_roi_zip(roi_zip_path[i])
            name = zip_list_all[i][k][:14]
            left_top = roi[name]['left'], roi[name]['top']
            right_bottom = left_top[0]+roi[name]['width'], left_top[1]+roi[name]['height']
            #print(left_top, right_bottom)        
            x1=str(left_top[0])
            y1=str(left_top[1])
            x2=str(right_bottom[0])
            y2=str(right_bottom[1])
            txt = '%s,%s,%s,%s,%s,diverticulitis\n'%(img_path, x1, y1, x2, y2)
            f.write(txt)
            c+=1
    a+=len(check)
    print(i,a, c)
f.close()
```
