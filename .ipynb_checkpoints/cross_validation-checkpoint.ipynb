```python
import random
import os,errno
import sys
import shutil


def getFileList():

    split_num = 5;

    path_dir1 = 'D:/appendicitis/data/Appendicitis_downsampling'
    path_dir2 = 'D:/appendicitis/data/Diverticulitis'
    file_list1 = os.listdir(path_dir1)
    file_list2 = os.listdir(path_dir2)

    fold1 = int(len(file_list1)/split_num)
    fold2 = int(len(file_list2)/split_num)


    appendicitis_list = []
    diverticulitis_list = []
    temp_list1 = []
    temp_list1 = file_list1

    temp_list2 = []
    temp_list2 = file_list2


    # Appendicitis
    for i in range(0, split_num-1):
        appendicitis_list.append([])
        for j in range(1, fold1):
            a = random.randrange(0, len(temp_list1)-1)
            src = temp_list1[a - 1]
            appendicitis_list[i].append(src)
            del temp_list1[a-1]
    appendicitis_list.append([])
    
    for i in temp_list1:
        appendicitis_list[split_num-1].append(i)

    print('appendicitis')
    for i in range(0,split_num):
        for j in appendicitis_list[i]:
            src = path_dir1+'/'+j
            idx = i+1
            dst = 'D:/appendicitis/data/test_down/exp%d/appendicitis'%idx +'/'+j
            shutil.copy(src,dst)

        for k in range(0, split_num):
            if k == i :
                pass
            else :
                for m in appendicitis_list[k]:
                    src2 = path_dir1 + '/' + m
                    idx = i + 1
                    dst2 = 'D:/appendicitis/data/train_down/exp%d/appendicitis' %idx + '/' + m
                    shutil.copy(src2, dst2)

    # diverticulitis
    for i in range(0, split_num-1):
        diverticulitis_list.append([])
        for j in range(1, fold2):
            a = random.randrange(0, len(temp_list2)-1)
            src = temp_list2[a - 1]
            diverticulitis_list[i].append(src)
            del temp_list2[a-1]
    diverticulitis_list.append([])
    for i in temp_list2:
        diverticulitis_list[split_num-1].append(i)

    print('diverticulitis')
    for i in range(0,split_num):
        for j in diverticulitis_list[i]:
            src = path_dir2+'/'+j
            idx = i + 1
            dst = 'D:/appendicitis/data/test_down/exp%d/diverticulitis'%idx +'/'+j
            shutil.copy(src,dst)

        for k in range(0, split_num):
            if k==i :
                pass
            else :
                for m in diverticulitis_list[k]:
                    src2 = path_dir2 + '/' + m
                    idx = i + 1
                    dst2 = 'D:/appendicitis/data/train_down/exp%d/diverticulitis' % idx + '/' + m
                    shutil.copy(src2, dst2)


def makeFolder():
    # make folder
    folder = []
    split_num =5
    for i in range(1, split_num+1):
        folder = []
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/train')
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/test')
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/train/exp%d' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/train/exp%d/diverticulitis' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/train/exp%d/no_lesion' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/test/exp%d' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/test/exp%d/diverticulitis' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/test/exp%d/no_lesion' % i)
        folder.append('D:/appendicitis/data/divertisulitis-no_lesion/npy/exp%d' % i)

    for j in range(len(folder)):
        try:
            os.makedirs(folder[j])
        except:
            print("folder1 Failed to create directory!!!!!")


if __name__ == "__main__":
    makeFolder()
    #getFileList()
```
